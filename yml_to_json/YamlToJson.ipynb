```python
def format_category_subcategory(json_data_services):
    cd= {}
    for cat in json_data_services:
        cat_name = cat["category"]
        cd[cat_name] ={"id": gen_uuid4(), "name": cat_name, "icon": "./data/img/category.png", "children": []}
    return cd

def add_subcategory_services(uncd):
    for cat in json_data_services:
        cat_name = cat["category"]
        sub_cat_name = cat["subcategory"]
        uncd[cat_name]["children"].append(
            {
                "id": gen_uuid4(), 
                "name": sub_cat_name, 
                "icon": "./data/img/subcategory.png", 
                "children": [
                    {
                        "name": list(i.keys())[0], 
                        "icon": f"./data/img/logo/{list(i.keys())[0]}.png",
                        "id": gen_uuid4(),
                        
                        "children":[ 
                            {"icon" : f"./data/img/cloudproviders/{list(i.keys())[0]}/{j['icon']}", "name": j["name"]} if ".png" in j["icon"] else j for j in i[list(i.keys())[0]]]
                    }
                    for i in cat["service"]
                ]  
            }
        )
    return uncd
    
```
